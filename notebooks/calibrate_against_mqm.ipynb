# Calibrate `logprobs-mt-qe` signals against human judgement

This notebook is a **stub**. It exists to map out the calibration workflow; the cells below are intentionally empty so contributors can wire in their own labelled data without inheriting any particular choice.

## Workflow

1. Score a labelled corpus (WMT MQM, MLQE-PE, internal post-edit logs, ...) with `qe/score_live.py` or `qe/score_json.py`.
2. Load the resulting raw signals alongside the human label.
3. Plot, correlate, and regress.
4. Ship the calibration as a new preset under `mt/presets/` or as a separate package.

Signals worth correlating (see [`docs/signals.md`](../docs/signals.md) for full definitions):

- `mean_logprob`, `min_logprob`, `perplexity_proxy`
- `composite_score`, `mean_prob`, `min_lexical_prob`
- `mean_top_entropy`, `mean_margin`
- `weak_share`, `ambiguous_share`
- Review-span count, language-script drift count

Targets: MQM total score, DA score, post-edit time, HTER, character-level edit distance.
